```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y , color = 'g')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("This is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y , color = 'g')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("This is my bar plot")
plt.grid()
```

```python
## To generate a 3D grap

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c ='#9B35C5', alpha = .9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
